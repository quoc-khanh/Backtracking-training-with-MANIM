```python

class last_text(Scene):
    def construct(self):
        congratulations_text = Text("Congratulations you have reached the destination", font_size=24, color=WHITE)
        congratulations_text.set_color(YELLOW)
        congratulations_text.set_stroke(color=BLUE, width=2)
        congratulations_text.move_to(ORIGIN).shift(UP * 1)

        Text1 = Text("Member Names: ").shift(UP * 1)
        Text2 = Text("Tran Dam Quoc Khanh (11223079)").shift(UP * 0.5)
        Text3 = Text("Nguyen Duy Hung (11222611)")
        Text4 = Text("Nguyen Trong Hung (11222582)").shift(DOWN * 0.5)
        Text5 = Text("Tran Ngoc Son (11225650)")
        Text6 = Text("Trinh Phuong Anh (11220655)").shift(DOWN * 1.5)
        Text1.move_to(ORIGIN).scale(0.4)
        Text2.next_to(Text1, DOWN * 0.5).scale(0.4)
        Text3.next_to(Text2, DOWN * 0.5).scale(0.4)
        Text4.next_to(Text3, DOWN * 0.5).scale(0.4)
        Text5.next_to(Text4, DOWN * 0.5).scale(0.4)
        Text6.next_to(Text5, DOWN * 0.5).scale(0.4)

        self.play(Write(congratulations_text))

        self.wait(3)
        for text in [Text1, Text2, Text3, Text4, Text5, Text6]:
            self.play(Write(text))
        self.wait(3)
```
